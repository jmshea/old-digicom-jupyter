```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, expon
%matplotlib inline
```

```python
def q(x):
    return norm.isf(x)
```

```python
ebno_db=np.linspace(0,10,11)
ebno=10**(ebno_db/10)
```

```python
ps_awgn = 1/2 * np.exp(-ebno)
plt.semilogy(ebno_db, ps_awgn)
```

```python
ps_rayleigh = 1/(2*ebno+2)
plt.semilogy(ebno_db, ps_rayleigh)
```

```python
ebno_db2=np.linspace(0,40,41)
ebno2 = 10** (ebno_db2/10)

ps_rayleigh = 1/(2*ebno2+2)
plt.semilogy(ebno_db2, ps_rayleigh)
plt.grid()
```

```python
# Simulation of just fading gains
num_sims = 100_000
ps = np.zeros(np.shape(ebno2))
E = expon()
Evals = E.rvs(size=num_sims)
for i, ebno in enumerate(ebno2):
    faded = Evals*ebno
    ps[i]+=np.mean(1/2*np.exp(-faded))

plt.semilogy(ebno_db2, ps);
plt.grid()
```
